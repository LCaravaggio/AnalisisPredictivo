# Loss Functions

# Regresi√≥n

```python
import numpy as np
```

```python
# MAE loss
def mae(t1, t2):
    diff = abs(t1 - t2)
    return diff 


# MSE loss
def mse(t1, t2):
    diff = t1 - t2
    return diff**2


# Hubber loss
def huber_loss(t1, t2, alpha):
  diff = abs(t1 - t2)
  if diff <= alpha: 
    return diff**2
  else:
    return alpha*diff - 0.5*alpha**2
```

```python
t1=[*range(-1000, 1000, 1)]
```

```python
error=[]
for e in t1: 
  error.append(mae(e,0))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
error=[]
for e in t1: 
  error.append(mse(e,0))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
t1=[*range(-10, 11, 1)]
```

```python
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 0.1))
e1=lista  
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 0.5))
e2=lista  
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 1))
e3=lista  
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 10))
e4=lista  
```

```python
import matplotlib.pyplot as plt
plt.plot(t1, e1, label = "Alpha =0.1")
plt.plot(t1, e2, label = "Alpha =0.5")
plt.plot(t1, e3, label = "Alpha =1")
plt.plot(t1, e4, label = "Alpha =10")
ax = plt.gca()
ax.set_ylim([0, 15])
ax.set_xlim([-10, 10])
plt.legend()
plt.show()
```

# Clasificaci√≥n

```python
def exponential_loss(y_pred, y_true):
    return np.mean(np.exp(- y_pred * y_true))

def Hinge(y_pred, y_true):
    return np.max([0., 1. - y_pred * y_true])

def kl_divergence(y_true, y_pred):
    return y_true * np.log(y_true / y_pred)

def cross_entropy(predictions, targets, epsilon=1e-10):
    ce_loss = -np.sum(np.sum(targets * np.log(predictions + 1e-5)))
    return ce_loss
```

```python
t1=[*range(-2, 4, 1)]
error=[]
for e in t1: 
  error.append(exponential_loss(e,1))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
t1=[*range(-2, 4, 1)]
error=[]
for e in t1: 
  error.append(Hinge(e,1))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
t1=[*range(-2, 4, 1)]
error=[]
for e in t1: 
  error.append(kl_divergence(e,1))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
t1=[*range(-2, 4, 1)]
error=[]
for e in t1: 
  error.append(cross_entropy(e,1))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```
