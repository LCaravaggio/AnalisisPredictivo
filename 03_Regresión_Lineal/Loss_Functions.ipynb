# Loss Functions

```python
import numpy as np
```

```python
# MAE loss
def mae(t1, t2):
    diff = abs(t1 - t2)
    return diff 


# MSE loss
def mse(t1, t2):
    diff = t1 - t2
    return diff**2


# Hubber loss
def huber_loss(t1, t2, alpha):
  diff = abs(t1 - t2)
  if diff <= alpha: 
    return diff**2
  else:
    return alpha*diff - 0.5*alpha**2
```

```python
t1=[*range(-1000, 1000, 1)]
```

```python
error=[]
for e in t1: 
  error.append(mae(e,0))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
error=[]
for e in t1: 
  error.append(mse(e,0))
```

```python
import matplotlib.pyplot as plt
plt.plot(t1,error)
```

```python
t1=[*range(-10, 11, 1)]
```

```python
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 0.1))
e1=lista  
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 0.5))
e2=lista  
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 1))
e3=lista  
lista=[]
for e in t1: 
  lista.append(huber_loss(e,0, alpha = 10))
e4=lista  
```

```python
import matplotlib.pyplot as plt
plt.plot(t1, e1, label = "Alpha =0.1")
plt.plot(t1, e2, label = "Alpha =0.5")
plt.plot(t1, e3, label = "Alpha =1")
plt.plot(t1, e4, label = "Alpha =10")
ax = plt.gca()
ax.set_ylim([0, 15])
ax.set_xlim([-10, 10])
plt.legend()
plt.show()
```
